# Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

# Generate Data

```python
def get_data(number_of_oberservations):
  """
  
  """
  # x-axis data
  x = np.linspace(start=-5, stop=5, num=number_of_oberservations)
  # generate y-axis/label using a straight line formula
  # slope = 2
  # bias = -1
  y = 2*x-1
  return x, y
```

```python
x,y = get_data(number_of_oberservations=11)
x,y
```

# Neural Network - Regressor

```python
model = tf.keras.Sequential(layers = [
                                      tf.keras.layers.Input(shape=(1,)),
                                      tf.keras.layers.Dense(units=1, use_bias=True)
])
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss='mse', metrics=["mean_squared_error"])
```

```python
# Train on the data
model.fit(x=x, y=y, epochs=500, batch_size = 1)
```

# Evaluate

```python
# predict for new value
model.predict([10])
```

```python
# plot the actual and predicted values
fig, axs = plt.subplots(figsize = (20, 10))
axs.plot(x, y, label='actual', color ='blue',marker ='o')
axs.plot(x, model.predict(x), label='predicted', color='red', marker="*")
axs.set_xlabel('input')
axs.set_ylabel('output')
axs.set_title("Neural Network regressor")
axs.legend()
```
