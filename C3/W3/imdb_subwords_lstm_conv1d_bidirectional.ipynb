# Nueral Network

```python
train_sequences = subwords_imdb['train']
test_sequences = subwords_imdb['test']
```

```python
type(train_sequences)
```

```python
train_sequences = train_sequences.padded_batch(batch_size=32)
test_sequences = test_sequences.padded_batch(batch_size=32)
```

```python
model = tf.keras.Sequential(
    layers = [
              tf.keras.layers.Input(shape=(None,)),
              tf.keras.layers.Embedding(input_dim=subwords_tokenizer.vocab_size, output_dim=64),
              tf.keras.layers.GlobalAveragePooling1D(),
              tf.keras.layers.Dense(units=6, activation='relu'),
              tf.keras.layers.Dense(units=1, activation='sigmoid')
    ]
)
model.summary()
```

```python
# compile
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
# train
model.fit(train_sequences, epochs=10, validation_data=test_sequences)
```

# LSTM

```python
model = tf.keras.Sequential(
    layers = [
              tf.keras.layers.Input(shape=(None,)),
              tf.keras.layers.Embedding(input_dim=subwords_tokenizer.vocab_size, output_dim=64),
              tf.keras.layers.LSTM(units=64),
              tf.keras.layers.Dense(units=64, activation='relu'),
              tf.keras.layers.Dense(units=1, activation='sigmoid')
    ]
)
model.summary()
```

```python
# compile
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
# train
model.fit(train_sequences, epochs=10, validation_data=test_sequences)
```

# Bidirectional LSTM

```python
model = tf.keras.Sequential(
    layers = [
              tf.keras.layers.Input(shape=(None,)),
              tf.keras.layers.Embedding(input_dim=subwords_tokenizer.vocab_size, output_dim=64),
              tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=64)),
              tf.keras.layers.Dense(units=64, activation='relu'),
              tf.keras.layers.Dense(units=1, activation='sigmoid')
    ]
)
model.summary()
```

```python
# compile
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
# train
model.fit(train_sequences, epochs=10, validation_data=test_sequences)
```

# Bidirectional stacked LSTM

```python
model = tf.keras.Sequential(
    layers = [
              tf.keras.layers.Input(shape=(None,)),
              tf.keras.layers.Embedding(input_dim=subwords_tokenizer.vocab_size, output_dim=64),
              tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=64, return_sequences=True)),
              tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=32)),
              tf.keras.layers.Dense(units=64, activation='relu'),
              tf.keras.layers.Dense(units=1, activation='sigmoid')
    ]
)
model.summary()
```

```python
# compile
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
# train
model.fit(train_sequences, epochs=10, validation_data=test_sequences)
```

# Conv1D

```python
model = tf.keras.Sequential(
    layers = [
              tf.keras.layers.Input(shape=(None,)),
              tf.keras.layers.Embedding(input_dim=subwords_tokenizer.vocab_size, output_dim=64),
              tf.keras.layers.Conv1D(filters=128, kernel_size=5),
              tf.keras.layers.GlobalAveragePooling1D(),
              tf.keras.layers.Dense(units=64, activation='relu'),
              tf.keras.layers.Dense(units=1, activation='sigmoid')
    ]
)
model.summary()
```

```python
# compile
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
# train
model.fit(train_sequences, epochs=10, validation_data=test_sequences)
```
